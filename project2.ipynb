```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import inspect
```

### Store CSV into DataFrame

```python
csv_file_1 = "Resources/country_profile_variables.csv"
cpv_df = pd.read_csv(csv_file_1)
cpv_df.head()
```

```python
csv_file_2 = "Resources/height_of_male_and_female_by_country.csv"
hmfc_df = pd.read_csv(csv_file_2)
hmfc_df.info()
```

```python
csv_file_3 = "Resources/world_bmi.csv"
wbmi_df = pd.read_csv(csv_file_3)
wbmi_df.info()
```

### Create new data with select columns

```python
new_cpv_df = cpv_df.drop(['Urban population growth rate (average annual %)', 'Fertility rate, total (live births per woman)', 'Life expectancy at birth (females/males, years)','International migrant stock (000/% of total pop.)', 'Refugees and others of concern to UNHCR (in thousands)','Infant mortality rate (per 1000 live births','Health: Physicians (per 1000 pop.)','Education: Government expenditure (% of GDP)','Education: Primary gross enrol. ratio (f/m per 100 pop.)','Education: Secondary gross enrol. ratio (f/m per 100 pop.)','Education: Tertiary gross enrol. ratio (f/m per 100 pop.)','Seats held by women in national parliaments %','Mobile-cellular subscriptions (per 100 inhabitants)', 'Mobile-cellular subscriptions (per 100 inhabitants).1','Individuals using the Internet (per 100 inhabitants)','Threatened species (number)','Forested area (% of land area)','CO2 emission estimates (million tons/tons per capita)','Energy production, primary (Petajoules)','Energy supply per capita (Gigajoules)','Pop. using improved drinking water (urban/rural, %)','Pop. using improved sanitation facilities (urban/rural, %)','Net Official Development Assist. received (% of GNI)'], axis=1)
new_cpv_df.insert(0, 'id', range(1, 1 + len(new_cpv_df)))
new_cpv_df.head()
```

```python
hmfc_df.insert(0, 'id', range(1, 1 + len(hmfc_df)))
hmfc_df.head()
```

```python
wbmi_df.insert(0, 'id', range(1, 1 + len(wbmi_df)))
wbmi_df.head()
```

```python
pi_merged_df = pd.merge(hmfc_df, wbmi_df, on=["id"])
pi_merged_df.drop(['Country', 'Both'], axis=1, inplace=True)
pi_merged_df.rename(columns={'Male': 'Boddy mass Male (kg/m2)', 'Female': 'Boddy mass Female (kg/m2)'}, inplace=True)
pi_merged_df.head()
```

### Connect to local database

```python
protocol = 'postgresql'
username = 'postgres'
password = 'postgres'
host = 'localhost'
port = 5432
database_name = 'test3'
rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string)
insp = inspect(engine)
```

### Check for tables

```python
insp.get_table_names()
```

### Use pandas to load csv converted DataFrame into database

```python
new_cpv_df.to_sql(name='country_profile_variables', con=engine, if_exists='append', index=False)
```

### Use pandas to load json converted DataFrame into database

```python
pi_merged_df.to_sql(name='physical_indicators', con=engine, if_exists='append', index=False)
```
